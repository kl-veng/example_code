```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('positions_df.csv')
df['position'] = df['position'].apply(ast.literal_eval)
```

```python
df['position'] = df['position'].apply(lambda lst: [word.replace('ſ', 's') for word in lst])
df['pub_date'] = pd.to_datetime(df['pub_date'].astype(str), format='%Y%m%d')
```

```python
df
```

Get Absolute Frequencies for a given position and display them over time

```python
searched_position = 'Verkäufer'
```

```python
count = df['position'].apply(lambda x: searched_position in x if isinstance(x, list) else False).sum()
print(count)
```

```python
position_df = df[df['position'].apply(lambda x: searched_position in x if isinstance(x, list) else False)]
frequency_per_year = position_df.groupby(position_df['pub_date'].dt.year).size()
```

```python
plt.figure(figsize=(10, 5))
frequency_per_year.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title(f'Frequency of {searched_position} Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

Get Relative Frequencies for a given position and display them over time

```python
searched_position = 'Verkäufer'
```

```python
# Group by year
total_ads_per_year = df.groupby(df['pub_date'].dt.year).size()
position_ads_per_year = df[df['position'].apply(lambda x: searched_position in x if isinstance(x, list) else False)].groupby(df['pub_date'].dt.year).size()

# Calculate relative frequencies
relative_frequencies = (position_ads_per_year / total_ads_per_year).fillna(0)  # Fill NaNs with 0
relative_frequencies = relative_frequencies[relative_frequencies > 0]
```

```python
# Create the plot
plt.figure(figsize=(10, 5))
relative_frequencies.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Relative Frequency')
plt.title(f'Relative Frequency of "{searched_position}" in Ads Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

Compare the position in job offers and job searches

```python
searched_position = 'Mädchen für alles'.lower() # Convert the word to lowercase if required
```

```python
# Filter by job_search and job_offer labels
job_search_ads = df[df['label'] == 'job_search']
job_offer_ads = df[df['label'] == 'job_offer']

# Filter for the given position in the 'position' column, ignoring case
job_search_hits = job_search_ads[job_search_ads['position'].apply(lambda x: searched_position in [w.lower() for w in x] if isinstance(x, list) else False)]
job_offer_hits = job_offer_ads[job_offer_ads['position'].apply(lambda x: searched_position in [w.lower() for w in x] if isinstance(x, list) else False)]

# Group by year and count the occurrences
job_search_per_year = job_search_hits.groupby(job_search_hits['pub_date'].dt.year).size()
job_offer_per_year = job_offer_hits.groupby(job_offer_hits['pub_date'].dt.year).size()

# Align the data for both categories to ensure both have the same years, filling missing values with 0
all_years = sorted(set(job_search_per_year.index).union(set(job_offer_per_year.index)))
job_search_per_year = job_search_per_year.reindex(all_years, fill_value=0)
job_offer_per_year = job_offer_per_year.reindex(all_years, fill_value=0)
```

```python
# Create the plot
plt.figure(figsize=(10, 5))

# Define the width of bars and the positions
bar_width = 0.4
indices = np.arange(len(all_years))

# Bar plot for job_search and job_offer
plt.bar(indices - bar_width/2, job_search_per_year.values, width=bar_width, color='blue', label='Job Search')
plt.bar(indices + bar_width/2, job_offer_per_year.values, width=bar_width, color='orange', label='Job Offer')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Absolute Frequency')
plt.title(f'Absolute Frequency of "{searched_position}" in Job Ads Over Time')
plt.xticks(indices, all_years, rotation=45)  # Rotate x-axis labels for better readability
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

Compare frequencies of three different positions for both genders

```python
def count_position_hits(df, position):
    count = df['position'].apply(lambda x: position in x if isinstance(x, list) else False).sum()
    return count
```

```python
position1 = 'Verkäufer'
position2 = 'Verkäuferin'
position3 = 'Buchhalter'
position4 = 'Buchhalterin'
position5 = 'Koch'
position6 = 'Köchin'
```

```python
positions_names = [position1, position2, position3, position4, position5, position6]
positions_frequencies = [count_position_hits(df, position) for position in positions_names]
```

```python
colors = ['blue', 'blue', 'orange', 'orange', 'green', 'green']

# Create the bar plot
plt.bar(positions_names, positions_frequencies, color=colors)

# Adding labels and title
plt.xlabel('Job Titles')
plt.ylabel('Counts')
plt.title('Job Title Counts')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
